```python
#Load in Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Data Cleaning
df = pd.read_csv('MDACUTE_Database.csv')
df.iloc[:,20:50]
```

```python
#This is creating a dictionary that maps the old column names to more appropriate coding counterparts

col_rename = ["SU1","SU2","SU3","SU4","SU8","SU6","SU5","SU9","SU10","SU7",
              "SU111","SU112","SU113","SU114","SU115","SU116",
              "SU121","SU122","SU123","SU124","SU125","SU126",
              "SU131","SU132","SU133","SU134","SU135","SU136",
              "SU141","SU142","SU143","SU144","SU145","SU146",
              "SU15"] #New column names

len(col_rename)
```

```python


colids = [] #
for i in (list(range(255,290))):
    colids.append(df.columns[i])

map = {}
for key in colids:
    for value in col_rename:
        map[key] = value
        col_rename.remove(value)
        break

df.rename(columns=map,inplace=True)
```

```python
df.rename(columns={"RSSCALE":"rsscore"}, inplace=True)
```

```python
List = [list(range(20)),list(range(34,300)),list(range(364,417)),418]
cols = []
for i in List:
    if type(i) == type(List):
        for j in i:
            cols.append(j)
    else:
        cols.append(i)
```

```python
#df = df[(df['Eligibility_QTotal'] > 0) & (df['Group'] != 0)]
#df = df['scr']
df = df.iloc[:,cols]
```
